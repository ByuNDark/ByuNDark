```python
import pandas as pd
from IPython.display import Image, display, HTML
import ipywidgets as widgets
from PIL import Image

photo_directory = "C:/Users/unkno/OneDrive/바탕 화면/incheon/visions"
def get_player_photo(name):
    return f"{photo_directory}/{name}.jpg"

matches = pd.read_csv('C:/Users/unkno/OneDrive/바탕 화면/incheon/matches.csv')
race = pd.read_csv('C:/Users/unkno/OneDrive/바탕 화면/incheon/race.csv')
sex = pd.read_csv('C:/Users/unkno/OneDrive/바탕 화면/incheon/sex.csv')

P = [];Q = [];R = [];S = []

for i in range(len(matches)):
    if (race['Protoss'] == matches['winner'][i]).sum() == 1:
        P.append('Protoss')
    elif (race['Terran'] == matches['winner'][i]).sum() == 1:
        P.append('Terran')
    else:
        P.append('Zerg')
    
    if (sex['남성'] == matches['winner'][i]).sum() == 1:
        Q.append('남성')
    else:
        Q.append('여성')
        
    if (race['Protoss'] == matches['loser'][i]).sum() == 1:
        R.append('Protoss')
    elif (race['Terran'] == matches['loser'][i]).sum() == 1:
        R.append('Terran')
    else:
        R.append('Zerg')
    
    if (sex['남성'] == matches['loser'][i]).sum() == 1:
        S.append('남성')
    else:
        S.append('여성')

PQdf = pd.DataFrame(data = [P, Q], index = ['win_종족', 'win_성별']).T
RSdf = pd.DataFrame(data = [R, S], index = ['lose_종족', 'lose_성별']).T

matches = pd.concat([matches, PQdf, RSdf], axis = 1)
MainDataField = matches[['winner','win_종족','win_성별','loser','lose_종족','lose_성별','map',
                         '리마스터','방식','대회 레벨','대회 이름','중요도']]

def SangDaeJunJuk(A, B): # A랑 B에 이름 작성할 때, csv에 있는 양식과 똑.같.이. 복붙해야 하며, 따옴표 없이 실행하면 오류 발생하므로 주의
    three_step = [];a = 0;b = 0
    one_step = MainDataField.index[MainDataField['winner'] == A].tolist()
    for i in one_step:
        if MainDataField.loc[i][3] == B:
            three_step.append(i)
            a += 1
    two_step = MainDataField.index[MainDataField['winner'] == B].tolist()
    for j in two_step:
        if MainDataField.loc[j][3] == A:
            three_step.append(j)
            b += 1
    result = sorted(three_step)
    
    try:
        photo_A = get_player_photo(A)
        photo_B = get_player_photo(B)

        # 이미지 불러오기
        image_A = widgets.Image(value=open(photo_A, 'rb').read(), width=150, height=150)
        image_B = widgets.Image(value=open(photo_B, 'rb').read(), width=150, height=150)

        # 텍스트 생성
        text = '{} {} : {} {}'.format(A, a, b, B)

        # 텍스트 위젯 (중간에 배치)
        text_widget = widgets.Label(value=text)
        text_widget.style = {'font_size': '25px'}

        # 수평 박스에 이미지와 텍스트 추가
        left = widgets.VBox([image_A])
        right = widgets.VBox([image_B])
        center = widgets.VBox([text_widget],layout=widgets.Layout(align_self='center'))

        # HBox를 사용해 좌우 이미지를 배치하고, 가운데 텍스트를 맞추기
        row = widgets.HBox([left, center, right])
        display(row)
    except FileNotFoundError:
        print('{} {} : {} {}'.format(A, a, b, B))

    return MainDataField.loc[result]

def MapData(C): # C에 이름 작성할 때, csv에 있는 양식과 똑.같.이. 복붙해야 하며, 따옴표 없이 실행하면 오류 발생하므로 주의
    result = MainDataField.index[MainDataField['map'] == C].tolist()
    return MainDataField.loc[result]

def PlayerData(D):
    one_step = MainDataField.index[MainDataField['winner'] == D].tolist()
    two_step = MainDataField.index[MainDataField['loser'] == D].tolist()
    final_step = sorted(one_step + two_step)
    
    photo_D = get_player_photo(D)
    
    try:
        display(widgets.Image(value=open(photo_D, 'rb').read(), width=150, height=150))
    except FileNotFoundError:
        print(f"사진 파일을 찾을 수 없습니다: {photo_D}")
    
    return MainDataField.loc[final_step]
```
