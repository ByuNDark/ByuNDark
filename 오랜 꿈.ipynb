```python
# 주사위 1개 시행(Turn = 0인 상태)
# 1 ~ 3 : 1P가 먼저 표식 깔기
# 4 ~ 6 : 2P가 먼저 표식 깔기
# 게임 시작(과 동시에 Turn = 1)
# while 누군가 (내 점수 = target_number) 달성:
#     1P 자원증감단계
#     1P 턴 소모
#     1P -> 2P 넘어가기 전 단계(ex : 땅거미 지뢰, 캐릭터 별 개인 기술, ...)
#     2P 자원증감단계
#     2P 턴 소모
#     2P -> 1P 넘어가기 전 단계(ex : 땅거미 지뢰, 캐릭터 별 개인 기술, ...)
#     Turn += 1 하기 전 단계(ex : 맵 상에 있는 특수 구조물, 캐릭터 별 개인 기술, ...)
# target_number 달성한 사람 win!
```

```python
# import 재료
import random
import math
import time
```

```python
print('게임 시작 준비')
time.sleep(0.5)
Target_Number = int(input('Target number를 설정하세요(Integer 필수)'))
time.sleep(0.5)
Do = input('주사위 1개 시행(Press Any Key)')
time.sleep(0.5)
dice = random.randint(1, 6)
if dice <= 3:
    print('{}이므로 1Player가 표식을 먼저 설치'.format(dice))
else:
    print('{}이므로 2Player가 표식을 먼저 설치'.format(dice))
Turn = 1;Player1_Score = 0;Player2_Score = 0
print('게임 시작')
print('현재 턴 : {}'.format(Turn))
while ((Player1_Score < Target_Number) | (Player2_Score < Target_Number)):
    Do = input('주사위 2개 시행(Press Any Key)')
    time.sleep(0.5)
    dice_1 = random.randint(1, 6);dice_2 = random.randint(1, 6)
    dice_sum = dice_1 + dice_2
    print('1Player 자원 증감 단계 결과는 {}'.format(dice_sum))
    Do = input('(Press Any Key)')
    time.sleep(0.5)
    print('1Player 턴 소모')
    time.sleep(0.5)
    Player1_Score += random.randint(1, 3)
    print('1P 현재 점수 : {}'.format(Player1_Score))
    if Player1_Score >= Target_Number:
        break
    time.sleep(0.5)
    print('1P -> 2P 가기 전 단계(땅거미 지뢰라던가, 캐릭터 개인 기술이라던가...)')
    time.sleep(0.5)
    Do = input('주사위 2개 시행(Press Any Key)')
    time.sleep(0.5)
    dice_1 = random.randint(1, 6);dice_2 = random.randint(1, 6)
    dice_sum = dice_1 + dice_2
    print('2Player 자원 증감 단계 결과는 {}'.format(dice_sum))
    Do = input('(Press Any Key)')
    time.sleep(0.5)
    print('2Player 턴 소모')
    time.sleep(0.5)
    Player2_Score += random.randint(1, 3)
    print('2P 현재 점수 : {}'.format(Player2_Score))
    if Player2_Score >= Target_Number:
        break
    time.sleep(0.5)
    print('2P -> 1P 가기 전 단계(땅거미 지뢰라던가, 캐릭터 개인 기술이라던가...)')
    time.sleep(0.5)
    print('Turn += 1 하기 전 단계(ex : 맵 상에 있는 특수 구조물, 캐릭터 별 개인 기술, ...)')
    Turn += 1
    print('현재 턴 : {}'.format(Turn))
print('경기 끝났구요, 현재 턴은 {}입니다.'.format(Turn))
if Player1_Score >= Target_Number:
    print('1P 승리')
else:
    print('2P 승리')
```
