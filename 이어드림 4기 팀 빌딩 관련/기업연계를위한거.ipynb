```python
import random
import pandas as pd
```

```python
INU = pd.read_csv('C:/Users/김태성/Desktop/파이썬용CSV/실험용기업연계.csv')
INU
```

```python
# pandas에서 각 row마다 개별 list로 바꾸기
A = [];B = [];C = [];D = [];E = [];F = []
for x in range(len(INU)):
    if INU.loc[x][3] == '0':
        F.append(INU.loc[x].tolist()) # 설문 미제출자는 1,2,3지망 모두 '0'처리
    elif (INU.loc[x][3] == INU.loc[x][4]) & (INU.loc[x][4] == INU.loc[x][5]) & (INU.loc[x][3] != '0'):
        A.append(INU.loc[x].tolist()) # 1,2,3지망 모두 같음
    elif (INU.loc[x][3] == INU.loc[x][4]) & (INU.loc[x][3] != '0'):
        B.append(INU.loc[x].tolist())# 1,2지망 같음
    elif (INU.loc[x][3] == INU.loc[x][5]) & (INU.loc[x][3] != '0'):
        C.append(INU.loc[x].tolist()) # 1,3지망 같음
    elif (INU.loc[x][4] == INU.loc[x][5]) & (INU.loc[x][4] != '0'):
        D.append(INU.loc[x].tolist()) # 2,3지망 같음
    else:
        E.append(INU.loc[x].tolist()) # 모두 다름
```

## Normal version

```python
# random shuffle 다음 기업 배치, A는 지망 올인이므로 shuffle 의미 없음
# A -> B -> C -> D -> E -> F
w = 1;v = 0 # trigger의 switch 역할

while v == 0:
    기업들 = {
    'de기업1': [], 'de기업2': [], 'de기업3': [], 'de기업4': [], 'de기업5': [], 'de기업6': [], 'de기업7': [], 'de기업8': [], 'de기업9': [],
    'ds기업1': [], 'ds기업2': [], 'ds기업3': [], 'ds기업4': [], 'ds기업5': [], 'ds기업6': [], 'ds기업7': [], 'ds기업8': [], 'ds기업9': []
    } # while 반복을 대비한 list 초기화
    
    for z in A:
        key = z[3]
        if key in 기업들:
            기업들[key].append(z)
    # 각 리스트는 기업들['de기업1'], 기업들['ds기업3'] 등으로 접근
    
    random.shuffle(B)
    random.shuffle(C)
    random.shuffle(D)
    random.shuffle(E)

    for z in B: # 1 & 2지망 기업 배치 -> 자리 없으면 3지망 기업으로
        key = z[3]
        if key in 기업들 and len(기업들[key]) < 7:
            기업들[key].append(z)
        else:
            기업들[z[5]].append(z)

    for z in C: # 1 & 3지망 기업 배치 -> 자리 없으면 2지망 기업으로
        key = z[3]
        if key in 기업들 and len(기업들[key]) < 7:
            기업들[key].append(z)
        else:
            기업들[z[4]].append(z)

    for z in D: # 1지망 기업 배치 -> 자리 없으면 2 & 3지망 기업으로
        key = z[3]
        if key in 기업들 and len(기업들[key]) < 7:
            기업들[key].append(z)
        else:
            기업들[z[4]].append(z)

    for z in E: # 1지망 기업 배치 -> 자리 없으면 2지망 -> 또 자리 없으면 3지망
        key = z[3]
        if key in 기업들 and len(기업들[key]) < 7:
            기업들[key].append(z)
        else:
            key = z[4]
            if key in 기업들 and len(기업들[key]) < 7:
                기업들[key].append(z)
            else:
                기업들[z[5]].append(z)

    for z in F: # 설문 미제출자는 기업 팀들 중 팀 인원 적은 곳부터 채워나가기
        if z[2] == 'DE':
            DE_orb = [len(기업들['de기업1']),len(기업들['de기업2']),len(기업들['de기업3']),len(기업들['de기업4']),len(기업들['de기업5']),len(기업들['de기업6']),len(기업들['de기업7']),len(기업들['de기업8']),len(기업들['de기업9'])]
            기업들['de기업{}'.format(DE_orb.index(min(DE_orb))+1)].append(z)
        else:
            DS_orb = [len(기업들['ds기업1']),len(기업들['ds기업2']),len(기업들['ds기업3']),len(기업들['ds기업4']),len(기업들['ds기업5']),len(기업들['ds기업6']),len(기업들['ds기업7']),len(기업들['ds기업8']),len(기업들['ds기업9'])]
            기업들['ds기업{}'.format(DS_orb.index(min(DS_orb))+1)].append(z)

    # 각 팀 별 수준 점수 합계
    Topology = []
    for i in range(9):
        u = 0
        for j in 기업들['de기업{}'.format(i+1)]:
            u += j[1]
        Topology.append(u)

    for i in range(9):
        u = 0
        for j in 기업들['ds기업{}'.format(i+1)]:
            u += j[1]
        Topology.append(u)

    # 합격 / 불합격
    if any(15 > x for x in Topology) | any(y > 8 for y in DE_orb) | any(y > 8 for y in DS_orb):
        w += 1
    else:
        v = 1
    
for i in range(9):
    u = 0
    for j in 기업들['de기업{}'.format(i+1)]:
        u += j[1]
    print(기업들['de기업{}'.format(i+1)])
    print('de기업{}의 수준 점수 합 :'.format(i+1), u)
    print('\n')

for i in range(9):
    u = 0
    for j in 기업들['ds기업{}'.format(i+1)]:
        u += j[1]
    print(기업들['ds기업{}'.format(i+1)])
    print('ds기업{}의 수준 점수 합 :'.format(i+1), u)
    print('\n')

print('작동한 횟수 :', w)
```

# 특정 인원끼리 같이 가는 거 제거를 포함한 최종 version

```python
YDS = pd.read_csv('C:/Users/김태성/Desktop/파이썬용CSV/기업연계(트랙분리).csv')

# pandas에서 각 row마다 개별 list로 바꾸기
A = [];B = [];C = [];D = [];E = [];F = []
for x in range(len(YDS)):
    if YDS.loc[x][3] == '0':
        F.append(YDS.loc[x].tolist()) # 설문 미제출자는 1,2,3지망 모두 '0'처리
    elif (YDS.loc[x][3] == YDS.loc[x][4]) & (YDS.loc[x][4] == YDS.loc[x][5]) & (YDS.loc[x][3] != '0'):
        A.append(YDS.loc[x].tolist()) # 1,2,3지망 모두 같음
    elif (YDS.loc[x][3] == YDS.loc[x][4]) & (YDS.loc[x][3] != '0'):
        B.append(YDS.loc[x].tolist())# 1,2지망 같음
    elif (YDS.loc[x][3] == YDS.loc[x][5]) & (YDS.loc[x][3] != '0'):
        C.append(YDS.loc[x].tolist()) # 1,3지망 같음
    elif (YDS.loc[x][4] == YDS.loc[x][5]) & (YDS.loc[x][4] != '0'):
        D.append(YDS.loc[x].tolist()) # 2,3지망 같음
    else:
        E.append(YDS.loc[x].tolist()) # 모두 다름

# 팀 내 금지된 조합 확인 함수
def check_team_conflicts(team):
    # (1) 이동민과 전민지, 한보람, 정윤진은 같은 팀 불가
    if '이동민' in team and any(person in team for person in ['전민지', '한보람', '정윤진']):
        return True
    # (2) 이선영, 이유진, 유수연은 서로 같은 팀 불가
    if len(set(team).intersection(['이선영', '이유진', '유수연'])) > 1:
        return True
    return False

# 팀 배치 후 팀별 금지된 조합을 검사하고, 문제가 있으면 다시 배치
def assign_with_conflict_check(배정리스트, 기업들, 우선지망, 차선지망=None, 세번째지망=None):
    for z in 배정리스트:
        key = z[우선지망]  # 우선지망에 따라 배치 시도
        if key in 기업들 and len(기업들[key]) < 7:
            기업들[key].append(z)
            team_members = [person[0] for person in 기업들[key]]
            if check_team_conflicts(team_members):  # 금지된 조합 있는지 확인
                기업들[key].remove(z)  # 금지된 조합이면 팀에서 제외하고 재배치 시도
                if 차선지망:
                    assign_with_conflict_check([z], 기업들, 차선지망, 세번째지망=세번째지망)  # 차선지망으로 재배치
                elif 세번째지망:
                    기업들[z[세번째지망]].append(z)  # 3순위 지망으로 배치
        else:
            # 우선지망에 배치 불가능하면 차선지망으로 배치 시도
            if 차선지망 and (z[차선지망] in 기업들 and len(기업들[z[차선지망]]) < 7):
                기업들[z[차선지망]].append(z)
                team_members = [person[0] for person in 기업들[z[차선지망]]]
                if check_team_conflicts(team_members):
                    기업들[z[차선지망]].remove(z)
                    if 세번째지망:
                        assign_with_conflict_check([z], 기업들, 세번째지망)  # 3순위 지망으로 재배치
            elif 세번째지망:
                # 차선지망도 안되면 3순위 지망으로 배치 시도
                기업들[z[세번째지망]].append(z)

# 기존 로직에 각 팀 배치 시 금지된 조합 확인 로직 추가
w = 1
v = 0

while v == 0:
    기업들 = {
        'de기업1': [], 'de기업2': [], 'de기업3': [], 'de기업4': [], 'de기업5': [], 'de기업6': [], 'de기업7': [], 'de기업8': [], 'de기업9': [], 'de기업10': [],
        'ds기업1': [], 'ds기업2': [], 'ds기업3': [], 'ds기업4': [], 'ds기업5': [], 'ds기업6': [], 'ds기업7': [], 'ds기업8': [], 'ds기업9': [], 'ds기업10': []
    }
    
    # A는 동일 지망자들로 배치
    for z in A:
        key = z[3]
        if key in 기업들:
            기업들[key].append(z)
    
    random.shuffle(B)
    random.shuffle(C)
    random.shuffle(D)
    random.shuffle(E)
    
    # 각 리스트를 배치하면서 팀 조합 체크
    assign_with_conflict_check(B, 기업들, 3, 5)  # B 리스트 1 & 2지망
    assign_with_conflict_check(C, 기업들, 3, 4)  # C 리스트 1 & 3지망
    assign_with_conflict_check(D, 기업들, 3, 4)  # D 리스트 1지망
    assign_with_conflict_check(E, 기업들, 3, 4, 5)  # E 리스트 1지망
    
    # 설문 미제출자 처리
    for z in F:
        if z[2] == 'DE':
            DE_orb = [len(기업들[f'de기업{i+1}']) for i in range(10)]
            기업들[f'de기업{DE_orb.index(min(DE_orb))+1}'].append(z)
        else:
            DS_orb = [len(기업들[f'ds기업{i+1}']) for i in range(10)]
            기업들[f'ds기업{DS_orb.index(min(DS_orb))+1}'].append(z)

    # 각 팀의 수준 점수 합계를 확인하는 로직 (기존 코드 유지)
    Topology = []
    for i in range(10):
        u = sum(j[1] for j in 기업들[f'de기업{i+1}'])
        Topology.append(u)

    for i in range(10):
        u = sum(j[1] for j in 기업들[f'ds기업{i+1}'])
        Topology.append(u)

    # 합격/불합격 조건 확인 및 재시도 여부
    DE_orb = [len(기업들[f'de기업{i+1}']) for i in range(10)]
    DS_orb = [len(기업들[f'ds기업{i+1}']) for i in range(10)]
    
    if any(14 > x for x in Topology) or any(y_1 > 8 for y_1 in DE_orb) or any(y_2 > 8 for y_2 in DS_orb):
        w += 1
    else:
        v = 1

# 최종 팀별 배치 결과 출력 (기존 코드 유지)
for i in range(10):
    u = sum(j[1] for j in 기업들[f'de기업{i+1}'])
    team_members = ', '.join([f"({z[0]}, {z[1]})" for z in 기업들[f'de기업{i+1}']])
    print(f'de기업{i+1} : {team_members}')
    print(f'de기업{i+1}의 수준 점수 합 :', u)
    print('\n')

for i in range(10):
    u = sum(j[1] for j in 기업들[f'ds기업{i+1}'])
    team_members = ', '.join([f"({z[0]}, {z[1]})" for z in 기업들[f'ds기업{i+1}']])
    print(f'ds기업{i+1} : {team_members}')
    print(f'ds기업{i+1}의 수준 점수 합 :', u)
    print('\n')

print('작동한 횟수 :', w)

r = 0
for i in range(10):
    r += len(기업들[f'de기업{i+1}'])
for i in range(10):
    r += len(기업들[f'ds기업{i+1}'])
print('전체 인원 =', r)
```
